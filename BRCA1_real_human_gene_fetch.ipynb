```python
# Install and import
!pip install biopython

from Bio import Entrez, SeqIO

# Always identify yourself to NCBI
Entrez.email = "kiruthikak984@gmail.com"

# Fetch BRCA1 gene — real human cancer gene
handle = Entrez.efetch(
    db="nucleotide",
    id="NM_007294",
    rettype="gb",
    retmode="text"
)

record = SeqIO.read(handle, "genbank")
handle.close()

# Explore it
print(f"Gene Name: {record.name}")
print(f"Description: {record.description}")
print(f"Sequence Length: {len(record.seq)} base pairs")
print(f"First 100 bases: {record.seq[:100]}")
print(f"Number of features annotated: {len(record.features)}")
print("\n✅ You just pulled a real human gene from the global genome database.")
```
