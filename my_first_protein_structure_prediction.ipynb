```python
# We'll use ESMFold via API - no heavy installation needed
import requests
import json

# This is a real protein sequence — Human Insulin
# The protein that controls your blood sugar every day
sequence = "MALWMRLLPLLALLALWGPDPAAAFVNQHLCGSHLVEALYLVCGERGFFYTPKTRREAEDLQVGQVELGGGPGAGSLQPLALEGSLQKRGIVEQCCTSICSLYQLENYCN"

print(f"Protein sequence length: {len(sequence)} amino acids")
print(f"Sequence: {sequence}")
print("\nSending to ESMFold for 3D structure prediction...")

# Call ESMFold API
headers = {"Content-Type": "application/x-www-form-urlencoded"}
response = requests.post(
    "https://api.esmatlas.com/foldSequence/v1/pdb/",
    headers=headers,
    data=sequence,
    timeout=60
)

if response.status_code == 200:
    pdb_content = response.text

    # Save the structure file
    with open("insulin_structure.pdb", "w") as f:
        f.write(pdb_content)

    print("✅ Structure predicted successfully!")
    print(f"PDB file size: {len(pdb_content)} characters")
    print("\nFirst 3 lines of your protein structure file:")
    for line in pdb_content.split('\n')[:3]:
        print(line)
    print("\n🔬 You just predicted a real human protein structure using AI.")
else:
    print(f"API status: {response.status_code}")
    print("Try again in a moment — API may be busy")
```
